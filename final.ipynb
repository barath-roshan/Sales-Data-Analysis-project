```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("train.csv")


df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')
df['Month'] = df['Order Date'].dt.to_period('M')
df['Day'] = df['Order Date'].dt.date
df['Week'] = df['Order Date'].dt.to_period('W').astype(str)


total_sales = df['Sales'].sum()
total_orders = df['Order ID'].nunique()
print(f"Total Sales: ${total_sales:,.2f}")
print(f"Total Orders: {total_orders}")


product_sales = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=True)
top_10_products = product_sales.head(10)
bottom_10_products = product_sales.tail(10)


plt.figure(figsize=(10, 6))
top_10_products.plot(kind='barh', color='green')
plt.title('Top 10 Best-Selling Products')
plt.xlabel('Total Sales')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()


plt.figure(figsize=(10, 6))
bottom_10_products.plot(kind='barh', color='red')
plt.title('Bottom 10 Least-Selling Products')
plt.xlabel('Total Sales')
plt.tight_layout()
plt.show()


region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(8, 5))
region_sales.plot(kind='bar', color='orange')
plt.title('Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()

monthly_sales = df.groupby('Month')['Sales'].sum()
plt.figure(figsize=(12, 6))
monthly_sales.plot(kind='line', marker='o')
plt.title('Monthly Sales Over Time')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()


daily_sales = df.groupby('Day')['Sales'].sum()
plt.figure(figsize=(14, 5))
daily_sales.plot(linewidth=1)
plt.title('Daily Sales Trend')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()


weekly_sales = df.groupby('Week')['Sales'].sum()
plt.figure(figsize=(12, 5))
weekly_sales.plot(marker='o')
plt.title('Weekly Sales Trend')
plt.xlabel('Week')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


segment_sales = df.groupby('Segment')['Sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(6, 4))
segment_sales.plot(kind='bar', color='purple')
plt.title('Sales by Customer Segment')
plt.xlabel('Segment')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```
